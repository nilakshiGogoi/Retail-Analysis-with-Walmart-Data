```python
std_dev = []
cvar=[]
for i in range(1,45):
  dfs = dfg[dfg["Store"]==i]
  sd=dfs["Weekly_Sales"].std()
  std_dev.append(sd)
  data=dfs["Weekly_Sales"].to_numpy()
  cv = lambda x: np.std(x, ddof=1) / np.mean(x) * 100 
  d = cv(data)
  cvar.append(d)

  plt.figure(figsize=(8,300))
  plt.subplot(45,1,i)
  sns.histplot(dfs["Weekly_Sales"],kde=True,bins=30)
  plt.title("--------------------------- \nSales of Store "+str(i)+" \nCoefficient of mean to Standard deviation is "+str(d))
```

```python
a = std_dev.index(max(std_dev))
n=a+1
print("Store "+str(n)+" has the maximum standard deviation and the Coefficient of mean to Standard deviation is "+str(cvar[a]) )
```

```python
dfs1 = df[df["Store"]==1]
```

```python
x=dfs1.drop(["Weekly_Sales","Date"],axis=1)
y=dfs1["Weekly_Sales"]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=30)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(xtrain,ytrain)
```

```python
ypred = lr.predict(xtest)
```

```python
from sklearn.metrics import r2_score
r2_score(ytest,ypred)
```

```python
import statsmodels.api as sm
xsm= sm.add_constant(x)

#fit linear regression model
model = sm.OLS(y, xsm).fit()
#view model summary
print(model.summary())
```
